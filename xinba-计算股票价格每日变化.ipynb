```python
import pandas as pd
```

```python
df = pd.DataFrame(data = [['20230521','20230522','20230523','20230524','20230525']*2,\
                 ['股票A']*5+['股票B']*5,[11,14,12,15,17,21,24,22,24,23]],\
                 index = ['日期','名称','价格'])
df
```

```python
df1 = df.T
df1
```

```python
a = df1.groupby('名称')['价格'].transform('pct_change')
a
```
