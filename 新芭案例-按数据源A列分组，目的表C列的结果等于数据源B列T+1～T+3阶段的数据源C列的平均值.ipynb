```python
import pandas as pd
```

```python
df = pd.read_excel('需求-丙辰.xlsx',usecols = 'A:C',nrows = 18,skiprows = 1)
df
```

```python
def func(d):
    day = pd.offsets.Day(1)
    foo = d.apply(lambda x:d[d.B.between(day + x.B,day*3 + x.B)].C.mean(),axis =1)
    return foo
    
```

```python
d = df.groupby('A',group_keys = False).apply(func).fillna(0)
d
```

```python
df1 = df.assign(D=d)
df1
```
