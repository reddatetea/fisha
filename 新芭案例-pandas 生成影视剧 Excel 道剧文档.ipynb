```python
import pandas as pd
import faker  
f = faker.Faker('zh-cn')
```

```python
df = (
    pd.DataFrame()
    .assign(客户姓名 = [f.name() for i in range(10)])
    .assign(年龄 = [f.random_int(20,50) for i in range(10)])
    .assign(最后去电时间 = [f.date_between(start_date = '-1y',end_date = 'today').strftime('%Y年%m月%d日') 
                      for i in range(10)])
    .assign(意向 = [f.random_element(('有','无')) for i in range(10)])
    .assign(地址 = [f.address() for i in range(10)])
)
df
```
