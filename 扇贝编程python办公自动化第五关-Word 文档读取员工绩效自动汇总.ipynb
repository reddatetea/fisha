```python
# 使用 import 导入需要的库
import os, fnmatch, docx, openpyxl
# 文件存储目录
dir_path = r'D:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第五关\第五关课件\Desktop\绩效评价'

# 获得该目录下的所有 .docx 文件 
files = [file for file in os.listdir(dir_path) if fnmatch.fnmatch(file,'*.docx')]


# 创建列表，用于保存从 Word 文档中提取的信息
infos = []


# 建立循环，提取每个 Word 文档中的信息
for file in files:
  wp = docx.Document(os.path.join(dir_path,file))
  name = wp.paragraphs[0].runs[0].text
  department = wp.paragraphs[1].runs[4].text
  post = wp.paragraphs[1].runs[7].text
  pingfen = wp.tables[0].cell(6,1).text
  pingjiaren = wp.paragraphs[3].runs[1].text
  shenhei = wp.paragraphs[4].runs[1].text
  info = [name,department,post,pingfen,pingjiaren,shenhei]
  infos.append(info)
headers = ['姓名','部门','岗位','评分','评分人','审核人']
infos.insert(0,headers)

# 创建工作簿，将提取到的信息写入 Excel 表中
new_wb = openpyxl.Workbook()
ws = new_wb.active
for info in infos:
  ws.append(info)

# 将生成的工作簿保存到当前目录
# 请不要更改已经写好的 save() 语句
new_wb.save('绩效评价汇总.xlsx')
```

```python
import os
import openpyxl
import docx
import fnmatch
dir_path = r'D:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第五关\第五关课件\Desktop\员工信息'
files = [file for file in os.listdir(dir_path) if fnmatch.fnmatch(file,'*.docx')]
infos = []
for file in files:
    wd = docx.Document(os.path.join(dir_path,file))
    paras_with_index = list(enumerate(wd.paragraphs))
    for index,paras in paras_with_index :
        runs = [run.text for run in paras.runs ] 
        runs_with_index = list(enumerate(runs))
        for index1,content in runs_with_index :
            print(f'段落序号:{index}')
            print(f'样式序号:{index1}')
            print(f'样式内容:{content}')
        print(' ')
    name = wd.paragraphs[2].runs[1].text
    department = wd.paragraphs[4].runs[1].text
    post = wd.paragraphs[5].runs[1].text
    height = wd.tables[0].cell(0,1).text
    weight = wd.tables[0].cell(1,1).text
    value = wd.tables[0].cell(2,1).text
    info = [name,department,post,height,weight,value]
    infos.append(info)

header = ['姓名','部门','岗位','身高','体重','属性']
infos.insert(0,header)

new_wb = openpyxl.Workbook()
ws = new_wb.active
for info in infos:
    ws.append(info)
new_wb.save('员工信息汇总.xlsx')       

    
```
