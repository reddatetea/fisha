```python
from openpyxl import load_workbook
from openpyxl.chart import Reference, BarChart

wb = load_workbook(r'F:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第四关\第四关课件\Desktop\第四关练习\各部门渠道营收汇总.xlsx')

# 遍历 wb 中的所有工作表
for ws in wb.worksheets:
  print(f'正在为工作表{ws.title}绘制图表')

  # 初始化一个柱形图
  chart = BarChart()

  # 引用表格数据
  data = Reference(worksheet=ws, min_row=2, max_row=8, min_col=2, max_col=3)
  chart.add_data(data, from_rows=False, titles_from_data=True)

  # 引用“月份”列的单元格范围
  cats = Reference(worksheet=ws, min_row=3, max_row=8, min_col=1, max_col=1)
  # 将 cats 设置为类别轴的标签
  chart.set_categories(cats)

  # 设置图表标题
  chart.title = f"{ws.title}收支汇总"

  # 设置 x 轴的标题
  chart.x_axis.title = "月份"
  # 设置 y 轴的标题
  chart.y_axis.title = "金额"

  # 套用 34 号图表样式
  chart.style = 34

  # 将柱形图添加到工作表中，左上角与单元格 A10 对齐
  ws.add_chart(chart, 'A10')

wb.save('各部门收支汇总-插入柱形图.xlsx')
```

```python
from openpyxl import load_workbook
from openpyxl.chart import LineChart, Reference
wb = load_workbook(r'D:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第四关\第四关课件\Desktop\第四关练习\各地区玩偶销量汇总.xlsx')

# 遍历所有工作表
for ws in wb.worksheets:


  # 初始化一个折线图
  chart = LineChart()


  # 创建 Reference 对象，从表格引用数据
  data = Reference(worksheet = ws,min_row=3,max_row=6,min_col=1,max_col=6)
  

  # 使用数据 data 绘制折线
  chart.add_data(data, from_rows=True, titles_from_data=True)
  
  


  # 引用各季度的单元格范围
  cats = Reference(worksheet=ws,min_row=2,max_row=2,min_col = 2,max_col=5)

  # 将 cats 设置为 X 轴标签
  chart.set_categories(cats)
  
  


  # 设置 Y 轴标题
  chart.y_axis.title = '销量/万'

  # 设置图表标题
  chart.title = f'{ws.title}2020年玩偶销量'

  # 使用 34 号图表样式
  chart.style = '34'


  # 将图表添加到工作表
  ws.add_chart(chart,'a8')


# 保存工作簿到当前目录
# 请不要更改已经写好的 save() 语句
wb.save('上机作业.xlsx')
```
