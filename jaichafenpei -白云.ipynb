```python
import os
import numpy as np
import pandas as pd
import easygui
```

```python
fname = r'd:\a00nutstore\006\zw\baiyun\2020白云入库.xlsx'
sheetname = '2020'
df0 = pd.read_excel(fname,sheetname)
period_baiyun= easygui.enterbox('请输入期间"仿2022-02"', title=" 白云期间")
cols_0 =  df0.columns.to_list()       #原列名
cols = [i.strip() for i in cols_0]    #去空格后的列名
df0.columns =  cols                    #更名为去空格后的列名    
df0.rename(columns = {'白云\n期间':'period_baiyun','期间':'period'},inplace = True)

```

```python
df1 =df0.query('period_baiyun == @period_baiyun')
df2= df0.query('period_baiyun != @period_baiyun')
df1_1 = df1.loc[~df1['品名'].str.contains('返利|价差|冲减|多计|折扣')]
df1_2 = df1.loc[df1['品名'].str.contains('返利|价差|冲减|多计|折扣')]
jiner0 = df1_1['金额'].sum().round(2)
jiacha = df1_2.loc[df1_2['品名'].str.contains('价差')]["金额"].sum()
```

```python
price_dic = dict(zip(df1['类别'],df1['单价']) )
price_dic
```

```python
diff_dic1 = {'造三高松厚度卷筒纸':-450, '造三抄本卷筒纸': -450, '造三抄本卷筒纸-390': -550, '封面专用卷筒纸': -50}
diff_dic1
```

```python
else_con = df1_1['入库单号'].astype(str).isin(['1390','1391'])
df1_1_1 = df1_1.loc[~else_con]
df1_1_2 = df1_1.loc[else_con]
df1_1_2
```

```python
df1_1_1 = df1_1_1.assign(单价 = df1_1_1['单价']+df1_1_1['类别'].map(diff_dic1).fillna(0))
df1_11 = pd.concat([df1_1_1,df1_1_2])
               
```

```python
def newlei(df_):
    mask = ~(df_['仓库材料'].str.strip().str.endswith('390')) | (df_['仓库材料'].str.strip().str.endswith('370'))
    df_['newlei'] = df_['类别'].where(mask,'小卷筒纸')
    return df_
newlei(df1_11)   
```

```python
diff_dic2 = {'小卷筒纸':-200}
df1_11['newlei'].map(diff_dic2).fillna(0)
df1_11 = df1_11.assign(单价 = df1_11['单价']+df1_11['newlei'].map(diff_dic2).fillna(0))
df1_11['金额'] = (df1_11['单价']*df1_11['数量(吨)']).round(2)
df1_11['不含税金额'] =(df1_11['金额'] /1.13).round(2)
```

```python
jiner1 = df1_11['金额'] .sum()
jiner1
```

```python
jiner_diff = jiner1 - jiner0
jiner_diff
```

```python
df1_11.drop('newlei',axis = 1,inplace = True )
df22 = pd.concat([df1_11,df1_2])
df22
```
