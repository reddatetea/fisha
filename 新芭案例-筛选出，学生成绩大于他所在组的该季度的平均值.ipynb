```python
import pandas as pd
```

```python
df = pd.read_excel(r'https://www.gairuo.com/file/data/dataset/team.xlsx')
df.head()
```

```python
# 分组数据
mean_by_Q = df.groupby('team').mean(numeric_only = True)
mean_by_Q
```

```python
# 单个组
mean_by_Q.loc['A']
```

```python
# 按行编写应用计算
df.apply(lambda s: (s.loc['Q1':'Q4'] > mean_by_Q.loc[s.team]).all(),axis = 1)
```

```python
# 筛选
df[df.apply(lambda s: (s.loc['Q1':'Q4'] > mean_by_Q.loc[s.team]).all(),axis =1)]
```

```python
# 吉翾老师写的
(
 df.loc[
     df.filter(regex = 'Q').gt(df.filter(regex = 'team|Q').groupby('team').transform('mean')).all(axis=1)
 ]
)
```
