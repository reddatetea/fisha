```python
import easygui
import pandas as pd
import os
import re
```

```python
def shengcan(msg):
    fname = easygui.fileopenbox(msg)
    sheet_name = '工资'
    df= pd.read_excel(fname,sheet_name = sheet_name)
    df = df.loc[:,['车间','卡号','姓名','代扣社保']]
    df.columns = ['部门车间','卡号','姓名','社保']
    df = df.loc[df['社保'].notna()]
    df = df.loc[df['社保']==383]
    return df
```

```python
msg = '请选择生产人员工资"excel文件'
df_shengcan = shengcan(msg)
df_shengcan
```

```python
def xingguan(msg):
    fname = easygui.fileopenbox(msg)
    sheet_name = '工资'
    df= pd.read_excel(fname,sheet_name = sheet_name)
    df = df[df['公司'] == '双佳']
    df = df.loc[:,['部门','账号','姓名','社保']]
    df.columns = ['部门车间','卡号','姓名','社保']
    df = df.loc[df['社保'].notna()]
       
    return df
```

```python
msg = '请选择行管人员工资"excel文件'
df_xingguan = xingguan(msg)
df_xingguan
```

```python
fname_chebao = r"F:\a00nutstore\008\zw08\双佳社保\2024-03\3月养老.xls"
df_chebao = pd.read_excel(fname_chebao)
df_chebao = df_chebao.loc[:,['社会保障号码','姓名','个人合计']]
df_chebao
```
