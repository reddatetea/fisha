```python
lst0 = ['订单编号',
 '采购单号',
 '换货单的原始订单编号',
 '订单状态',
 '订单锁定状态',
 '供应商编号',
 '供应商商家名称',
 '供应商店铺名称',
 '分销商编号',
 '分销商商家名称',
 '分销商店铺名称',
 '运费',
 '收货人姓名',
 '联系方式',
 '收货地址',
 '订单备注',
 '订单创建时间',
 '订单出库时间',
 '订单完成时间',
 '订单取消时间',
 '是否京仓',
 '产品名称',
 '产品颜色',
 '产品尺码',
 '商家SKU',
 '父SKU',
 '子SKU',
 '产品采购价',
 '采购数量',
 '采购单支付状态',
 '采购单支付时间',
 '采购单支付模式',
 '采购单应付采购款',
 '采购单实际支付总额',
 '订单类型',
 '用户实际支付总额',
 '指定承运商',
 '物流运单号']
lst1 = ['订单编号',
 '订单状态',
  '订单创建时间',
  '产品尺码',
 '商家SKU',
 '产品采购价',
 '采购数量',
 '采购单应付采购款',
 '用户实际支付总额',
]
# def readexcel(fname):
#     df = pd.read_excel(fname)    #,dtype = {'订单编号':'str','商家SKU':'str', '订单创建时间':'datetime64[ns]'},engine = 'openpyxl'

fname = r"F:\a00nutstore\008\zw08\电商\京东\批量任务中心_18_18425067_订单生成时间20240901000000至20241125235959-1732604043310.xls"
df = pd.read_excel(fname,dtype = {'订单编号':'str','商家SKU':'str', '订单创建时间':'datetime64[ns]'})
df.columns = lst0
df =df[lst1]

```

```python
df['产品尺码'] = df['产品尺码'].fillna('1本装')
regex = re.compile(r'^(\d+)本装')  
def getShuliang(str):
    mat = re.search(regex,str)
    if mat:
        return int(mat.group(1))
    else :
        return 1
df1 =  df.assign(shuliang = df.apply(lambda x:getShuliang(x['产品尺码']),axis = 1))
df1.to_excel('京东销售明细原表1.xlsx',index = False)
   
```

```python
df1['采购单应付采购款'] = df1['采购单应付采购款'].fillna(method = 'ffill')
df2 = df1[~(df1['商家SKU'].isnull())]
df2 = df2[df2['订单状态'] == '已完成']
```

```python
df2.to_excel('京东分配前-3.xlsx',index = False)
```

```python
gp = df2.groupby('订单编号')
#将订单总金额按各产品的明细销售额进行分配
data = []
for k,v in gp:
    # print(k,v)
    v['数量'] = v['采购数量']*v['shuliang']
    v['jiner0'] = v['数量']*v['产品采购价']
    # mean = v['采购单应付采购款'].mean()
    total = v['jiner0'].sum()
    # print('mean',mean)
    # print('total',total)
    v['金额'] = v['jiner0']/total*v['采购单应付采购款']
    data.append(v)
result0 = pd.concat(data)
# result1 = result1.reset_index()

result0
```

```python
# result0.to_excel('jindong分配后.xlsx',index = False)
```

```python
regex1 = re.compile(r'(.*)-(\d+)$')  
def getCunhuobianma(str):
    mat = re.search(regex1,str)
    if mat:
        str1 = mat.group(1)
        str2 = int(mat.group(2))
        return str1,str2
    else :
        return str,1
result1 =  result0.assign(存货编码 = result0.apply(lambda x:getCunhuobianma(x['商家SKU'])[0],axis = 1))
result1 =  result1.assign(shuliang0 = result1.apply(lambda x:getCunhuobianma(x['商家SKU'])[1],axis = 1))

result1
```

```python
def getFinalShuliang(str1,str2):
    if str1 >= str2:
        return str1
    else :
        return str2
result1 =  result1.assign(shuliang8 = result1.apply(lambda x:getFinalShuliang(x['shuliang'],x['shuliang0']),axis = 1))
```

```python
result1['数量'] = result1['shuliang8'] *  result1['数量'] 
#按品名汇总
result2= pd.pivot_table(result1,index = '存货编码',values = ['数量','金额'],aggfunc = 'sum')
result2
```

```python
result3 = result2.copy()

result3
```

```python
result3['金额'] = result3['金额'].round(2)
result3 = result3.astype({'数量':int})
total =result3.sum().to_frame().T
total.name = '品名'
result3 = pd.concat([result3,total])
new_index = result3.index.to_list()
new_index[-1] = '合计'
result3.index = new_index

result3
```
