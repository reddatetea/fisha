需求：按照["Name", "Type"]分组，获取中位数，

并将price1，price2, price3中的每个值分别与分组后的中位数相对比。

如果>中位数，则标记为0， 否则为1，最后每行计算的0和1相加，

形成一个新列Value, Value列后续主要用于判断是否为异常值。

```python
import pandas as pd
```

```python
df = pd.read_clipboard()
```

```python
df
```

```python
# df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean, axis=None, numeric_only=True)
df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean, axis=None)  #has no numeric_only
df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean)  #has no numeric_only
```

```python
v = (
    (df.loc[:, 'Price1': 'Price3'] < df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean, axis=None))
    .astype(int)
    .sum(1)
)
v
```

```python
df.assign(Value=v)
```

```python
df.loc[:, 'Price1': 'Price3'] < df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean, axis=None)

```

```python
df1 = df.loc[:, 'Price1': 'Price3'] < df.groupby(['Name', 'Type']).transform(pd.DataFrame.mean, axis=None)
df1 = df1.astype(int)
v = df1.sum(1)
v
```

```python
df.assign(Value=v)
```
