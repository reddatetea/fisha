```python
import pandas as pd
from io import StringIO
```

```python
data1 = '''
name
Liver
Arry
Ack
Eorge
Acob
Lfie
Oscar
Leo
Logan
Thomas
'''
```

```python
data2 = '''
team    name  Q1  Q2
   A     Ack  57  60
   A    Lfie   9  10
   A   Oscar  77   9
   B    Acob  61  95
   B     Leo  17   4
   B   Logan   9  89
   B  Thomas  80  48
   C    Arry  36  37
   C   Eorge  93  96
   E   Liver  89  21
'''
```

```python
df1 = pd.read_csv(StringIO(data1))
df2 = pd.read_csv(StringIO(data2), delim_whitespace=True)
```

```python
df2.set_index('name')
```

```python
# 分组、拼接
(
    df2.set_index('name')
    .groupby('team')
    .pipe(lambda g: pd.concat([d for _,d in g], axis=1))
)
```

```python
# 将所有的 team 填充完整，并将缺失值修改为 0
(
    df2.set_index('name')
    .groupby('team')
    .pipe(lambda g: pd.concat([d for _,d in g], axis=1))
    .assign(team=lambda d: d.team.ffill().bfill() )
    .fillna(0)
)
```

```python
# 按 df1 提供的数据重新排序，并将 Q1 和 Q2 列的类型修改为整型
(
    df2.set_index('name')
    .groupby('team')
    .pipe(lambda g: pd.concat([d for _,d in g], axis=1))
    .assign(team=lambda d: d.team.ffill().bfill() )
    .fillna(0)
    .reindex(df1.name)
    .astype({col:int for col in ['Q1', 'Q2']})
)
```

```python
# 下面是吉翾老师写的
(
 pd.concat([
     df1
     .assign(team = x)
     .merge(df2,on = ['name','team'],how = 'left')
     .set_index('name')
     for x in df2.team.unique()
 ],axis = 1)
 .fillna(0)
 .astype({col:int for col in ['Q1','Q2']})
)
```
