```python
# 原数据
df = pd.DataFrame({
    'A':['a1', 'a1', 'a2', 'a2'],
    'B':['b1', 'b2', None, 'b2'],
    'C':[1, 2, 3, 4],
    'D':[5, 6, None, 8],
    'E':[5, None, 7, 8]
})

df


```

```python
# 将缺失值填充为0
df1 = df.fillna(0)
df1
```

```python
df
'''
    A     B  C    D    E
0  a1    b1  1  5.0  5.0
1  a1    b2  2  6.0  NaN
2  a2  None  3  NaN  7.0
3  a2    b2  4  8.0  8.0
'''


# 取后一个有效值填充
df.fillna(method='bfill')
'''
    A   B  C    D    E
0  a1  b1  1  5.0  5.0
1  a1  b2  2  6.0  7.0
2  a2  b2  3  8.0  7.0
3  a2  b2  4  8.0  8.0
'''

# 取前一个有效值填充
df.fillna(method='ffill')
'''
    A   B  C    D    E
0  a1  b1  1  5.0  5.0
1  a1  b2  2  6.0  5.0
2  a2  b2  3  6.0  7.0
3  a2  b2  4  8.0  8.0
'''
```

```python
fname = r"F:\repos\fish\填充.xlsx"
df = pd.read_excel(fname)
df
```

```python
df1 = df.copy()
df1
```

```python
group = df1.groupby('品名',)
group
```

```python
data = []
def fillPrice(df0):
    df = df0.sort_values('日期')
    df['单价'] = df['单价'].fillna(method='ffill').fillna(method='bfill')
    return df
for key,gp in group:
    df = fillPrice(gp)
    print(df)
    data.append(df)
data
    
```

```python
df8 = pd.concat(data)
df8.to_excel('1111.xlsx',index = False)
```
