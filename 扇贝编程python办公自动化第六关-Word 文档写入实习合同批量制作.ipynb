```python
import docx, openpyxl
wb = openpyxl.load_workbook(r'D:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第六关\第六关课件\Desktop\第六关练习\员工信息替换模板.xlsx')
ws = wb.active

# 定义文本替换函数
def replace_info(wb,old_info,new_info):
  for pada in wd.paragraphs:
    for run in pada.runs:
      run.text = run.text.replace(old_info,new_info)

# 读取 Excel 工作表中的第一行，作为被替换的旧信息 old_infos
old_infos = [cell.value for cell in ws[1]]

# 从第二行开始，按行读取 Excel 表中的数据
for i in range(2,ws.max_row+1):

  # 每读取一行，就从空白模板创建一个新的 Document 对象
  wd = docx.Document(r'D:\a00nutstore\fishc\扇贝编程python办公自动化\扇贝编程python办公自动化第六关\第六关课件\Desktop\第六关练习\员工信息登记 - 空白模板.docx')

  # 替换标识列表
  text = ['#姓名#', '#部门#', '#岗位#', '#身高#', '#体重#', '#属性#']
  # 设置替换模板，将替换标志添加到 Word 文档中
  for j in range(len(text)):
    wd.paragraphs[j+1].add_run(text[j])
    



  # 读取当前行每个单元格的信息，作为最终替换结果 new_infos
  new_infos = [cell.value for cell in ws[i]]
  # 让新旧信息一一对应
  replacelist = zip(old_infos,new_infos)


  # 调用 replace_info() 函数替换信息
  for old_info,new_info in replacelist:
    replace_info(wd,old_info,new_info)
    



  # 读取每行的第一个单元格，组成 Word 文档的名称，保存文档到当前目录
  # 请不要更改已经写好的 save() 语句
  wd.save(f'员工信息登记-{ws.cell(i,1).value}.docx')
```
