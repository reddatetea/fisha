```python
#莱新销货单

def saleDeliveryLaixin(df):
    
    df = df.dropna(how = 'all',axis = 0)
    cols = [i for i in df.columns.to_list() if 'Unnamed' not in i]
    df = df[cols]
    #get maxrows
    for label,ser in df.items():
     for num,x in enumerate(ser):
        if isinstance(x,str):
            if '合计' in x:
                index = num
                print(index)
                break
    df = df.iloc[:index]
    df = df.dropna(how = 'all',axis = 1)
    return df
# df  = saleDeliveryLaixin(df)
# df
```
