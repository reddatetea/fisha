```python
def getData():
    df = pd.read_clipboard()
    df = df.reset_index()
    df = df.iloc[:,[0,1,-2,-1]]
    df.columns = ['dep','hours','debit','credit']
    df = df.dropna(subset=['hours'])
    new_dep = df.dep.str.replace('Outlet 1','Riverside Café').replace('Outlet 2','Terrace').replace('Outlet 3','Crane Palace').replace('Bar 1','Toucan Bar').replace('Bar 2','Piano Bar')
    df.dep = new_dep
    df.dep = df.dep.ffill()
    df.debit = df.debit.fillna('').str.replace('-','')
    df.credit = df.credit.fillna('').str.replace('-','')
    df.debit = df.debit.str.strip()
    df.credit = df.credit.str.strip()
    return df
```

```python
df0 = getData()
df0
```

```python
df1 = df0.copy()
df1
```

```python
def getStyle(ser):
    newSer = []
    for i in ser:
        if i.startswith('('):
            j = i.replace('(','-')
            j = j.replace(')','')
            newSer.append(j)

        else :
            if i in [None,'']:
                newSer.append(i)

            else :
                j = '+' + i
                newSer.append(j)
    return newSer
                
    

```
