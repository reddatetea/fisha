```python
'''
2025-3-2聚水潭根据大库的调拨单转换为标准格式，分两次导入
1.聚水潭有编号的先导进去
2.没有编号的交电商运营补商品编码，如大库与聚水潭编码格式不一致，则先记入编码对照表
3.电商运营在聚水潭中补齐商品编码，并将对照表转过来后，再导入上面没有导入的部分
每20条记录分割为一个入库导入文件，将不能导入的商品编码复制到粘贴板,汇总后再导一次
'''
```

```python
import pandas as pd
import easygui
import os
import pyperclip
```

```python
#先核对纸质调拨单
fname = easygui.fileopenbox('请点选大库本次调拨单文件')
df = pd.read_excel(fname,dtype = {'存货编码':"str"})
path,_ = os.path.split(fname)
os.chdir(path)
df
```

```python
#对照表
fname_duizhaobiao = r"F:\a00nutstore\008\zww08\电商\聚水潭\聚水潭编码与大库差异.xlsx"
df_duizhaobiao = pd.read_excel(fname_duizhaobiao,dtype = {'jusuitan':'str','xiaoshoubu':'str'})
dic = dict(zip(df_duizhaobiao.xiaoshoubu,df_duizhaobiao.jusuitan))
dic
```

```python
df= df.loc[~df['存货编码'].isna()]
df = df.loc[(df['调出仓库'] == '发货仓库') & (df['调入仓库'] == '电商仓库')]
df = df[['存货编码','数量（本）']]
df = df.rename(columns = {'存货编码':'商品编码','数量（本）':'数量'})
gp = df.groupby('商品编码')['数量'].sum()
df1 = pd.DataFrame(gp)
df1['单价'] = 0
df1['备注'] = ''
# df2 = df1.copy()
def func(x):
    if x in dic:
        return dic.get(x)
    else :
        return x
df2 = df1.reset_index()
df2['商品编码'] = df2['商品编码'].apply(func)


```

```python
df2.to_excel('聚水潭入库模板.xlsx',index = False)
```

```python
df3 = df2.copy()
df3.insert(0,'序号',range(len(df3)))
df3
```

```python
lst = [i for i in range(0,len(df3),20)]
lst
```

```python
# pd.qcut(df3.序号,q =len(df3)//20 + 1)
```

```python
gp1 = df3.groupby(pd.qcut(df3.序号,q = len(df3)//20 + 1))
gp1
```

```python
#每20条数据分割后形成多个导入模板
def chuli(v):
    v = v.iloc[:,1:]
    return v
i = 0
for k,v in gp1:
    i = i+1
    j = str(i)
    v = chuli(v)
    v.to_excel(f'聚水潭导入模板{j}.xlsx',index = False)
```
