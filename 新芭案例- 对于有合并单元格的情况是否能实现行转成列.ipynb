```python
import pandas as pd
```

```python
df = pd.read_excel('行转成列.xlsx',sheet_name = 'Sheet1',usecols = 'A:C',names = [*'xyz'],header = None)
df
```

```python
df1 = (
 df.assign(x = df.x.ffill())
   .astype(str)
   .groupby(['x'])
   .agg(lambda m:m.to_list())
   .applymap(','.join)
)
df1
```

```python
df1  = df1.reset_index()
df1
```

```python
data = list(df1.values)
data
```

```python
import openpyxl
wb = openpyxl.load_workbook('行转成列.xlsx')
ws = wb['Sheet1']
myRow = 9
myColumn = 7
for i in range(myRow,len(data)+myRow):
    row = data[i-myRow]
    for j in range(myColumn,len(row)+myColumn):
        ws.cell(i,j).value = row[j-myColumn]
wb.save('行转成列.xlsx')
    
```

```python
import xlwings as xw
app = xw.App(visible = False,add_book = False)
wb = app.books.open('行转成列.xlsx')
ws = wb.sheets['Sheet1']
ws.range('G13').options(pd.DataFrame, index=False,header=None).value = df1
wb.save('行转成列.xlsx')
app.quit()


```
