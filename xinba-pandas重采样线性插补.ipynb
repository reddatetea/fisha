```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [10, 20, 50, 40, 80,],
                   'B': [2, 8, 10, 6, 4, ],
                  })

df
```

```python
df.index.repeat(3)
```

```python
df.loc[df.index.repeat(3)]
```

```python
(
    df.loc[df.index.repeat(3)]
    .iloc[:-3+1] # 删除最后三个（可为变量），再保留1个，方便以后封装
)
```

```python
def func(d: pd.DataFrame):
    d.iloc[1:, :] = None
    return d

(
    df.loc[df.index.repeat(3)]
    .iloc[:-3+1]
    .groupby(level=0)
    .apply(func)
)
```

```python
(
    df.loc[df.index.repeat(3)]
    .iloc[:-3+1]
    .groupby(level=0)
    .apply(func)
    .interpolate()
)
```
