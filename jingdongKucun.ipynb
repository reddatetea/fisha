```python
df1 = pd.read_excel(fname,dtype = {'订单编号':'str','商家SKU':'str', '订单创建时间':'datetime64[ns]'})
df1
```

```python
df1.columns = lst0
df1
```

```python
df2 =df1[lst1]
df2
```

```python
df2 = df2[df2['订单状态'] == '已完成']
```

```python
df2['产品尺码'] = df2['产品尺码'].fillna('1本装')
df2
```

```python
regex = re.compile(r'^(\d+)本装')  
def getDingdanhao(str):
    mat = re.search(regex,str)
    if mat:
        return int(mat.group(1))
    else :
        return 1
   
```

```python
df2 =  df2.assign(shuliang = df2.apply(lambda x:getDingdanhao(x['产品尺码']),axis = 1))
df2
```

```python
df3 = df2.copy()
df3['采购单应付采购款'] = df3['采购单应付采购款'].fillna(method = 'ffill')
df3
```

```python
df3.to_excel('1111.xlsx',index = False)
```

```python

df4 = df3[~(df3['商家SKU'].isnull())]
df4
```

```python
df4.to_excel('2222.xlsx',index = False)
```

```python
df4.to_excel('京东分配前.xlsx',index = False)
```

```python
gp = df4.groupby('订单编号')
#将订单总金额按各产品的明细销售额进行分配
data = []
for k,v in gp:
    # print(k,v)
    v['数量'] = v['采购数量']*v['shuliang']
    v['jiner0'] = v['数量']*v['产品采购价']
    # mean = v['采购单应付采购款'].mean()
    total = v['jiner0'].sum()
    # print('mean',mean)
    # print('total',total)
    v['金额'] = v['jiner0']/total*v['采购单应付采购款']
    data.append(v)
result1 = pd.concat(data)
# result1 = result1.reset_index()

result1
```

```python
result1.to_excel('jindong分配后.xlsx',index = False)
```

```python
#按品名汇总
pivot = pd.pivot_table(result1,index = '商家SKU',values = ['数量','金额'],aggfunc = 'sum')
pivot
```

```python
total =pivot.sum().to_frame().T
total.name = '品名'
result2 = pd.concat([pivot,total])

```
