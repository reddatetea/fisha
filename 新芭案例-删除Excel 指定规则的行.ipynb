```python
import pandas as pd
import xlwings as xw

wb = xw.Book('删除Excel 指定规则的行.xlsx')
sheet = wb.sheets[1]
sheet
```

```python
df = pd.DataFrame(sheet.used_range.value)
# rows = df[df.iloc[:, 0] == '合计'].index+1
rows = df[df.iloc[:, 0] == '小计'].index+1
rows
```

```python
# 重点：每批删除四行，但次删除会重新更新行号，需要重新计算
n = 0
for i in rows:
    i -= n*4       # i = i - n*4 每删除四行，减少四个行号
    print('i=',i)
    for _ in range(4):
        print('row=',_)
        sheet.range(f'{i}:{i}').delete()
        # print(i)
    n += 1    # 用来控制删除行数的次数，第一次为0，转到上面为0*4，第二次为1，转到上面为1*4
```

```python
wb.save('demo-done.xlsx')
wb.close()
```
