```python
import pandas as pd

data = [[1,2,3,4,5,False],
        [2,3,4,5,6,True],
        [3,4,5,6,7,False],
        [5,6,7,8,9,True],
        [6,7,8,9,0,False],
        [6,7,8,9,0,False],
        [6,7,8,9,0,False]]

df = pd.DataFrame(data, columns = ['a','b','c','d','e','f'])
df
```

```python
for i,v in df.f.items():
    print(i,v)
```

```python
def func(ser):
    lst = []
    value = 0
    for v in ser:
        if v == True:
            value = 0
        else :
            value += 1
        lst.append(value)
    return lst
         
        
```

```python
df.assign(g = df.f.pipe(func))
```

```python
df.assign(g = func(df.f))
```
