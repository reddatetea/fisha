```python
import pandas as pd
import numpy as np
```

```python
fname = r'team.xlsx'
df = pd.read_excel(fname)
df
```

```python
(
    df.groupby('team')
    .apply(lambda x:pd.Series({
        'Q1_sum':x['Q1'].sum(),
        'Q1_max':x['Q1'].max(),
        'Q1_mean':x['Q1'].mean(),
        'Q1_prodsum' : (x['Q4']*x['Q4']).sum()
    }))
)
```

```python
def f_mi(x):
    d =[]
    d.append(x.Q1.sum())
    d.append(x.Q2.sum())
    d.append(x.Q3.sum())
    d.append((x['Q4']*x['Q4']).sum())
    return pd.Series(d,index = [['Q1','Q2','Q3','Q4'],
                               ['sum','max','mean','prodsum']])
```

```python
df.groupby('team').apply(f_mi)
```

```python
df.groupby('team').pipe(lambda x : x.max() + x.min())
```

```python
def mean_diff(x):
    return x.get_group('A').mean() - x.get_group('B').mean()
df.groupby('team').pipe(mean_diff)
    
```
