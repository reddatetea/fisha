```python
import pandas as pd
```

```python
df = pd.read_clipboard()
df
```

```python
gp = df.replace('前',None).groupby('groupcode').agg(set)
gp
```

```python
df1 = df.replace('前', None)
df1
```

```python
df2 = df1.assign(tetx2 = df1.groupcode.map(gp.tetx))
df2
```

```python
df2 = df2.assign(tetx3 = df2.tetx2)
df2
```

```python
def fun(df2):
    if df2.tetx2 == {'左前','右前'}:
        df2.tetx3 = df2.tetx 
    else :
        if df2.tetx != None:
          df2.tetx3 = df2.tetx
        else :
          df2.tetx3 = list({'左前','右前',None} - df2.tetx2)[0]
    return df2
```

```python
df3 = df2.apply(fun,axis =1)
df3
```
