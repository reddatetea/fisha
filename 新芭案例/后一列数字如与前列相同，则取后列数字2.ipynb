```python
from io import StringIO
import pandas as pd
import numpy as np
```

```python
data = '''
user_id,2019-10,2019-11,2019-12,2020-01,2022-02
9794320,0,1,0,0,0
10079204,0,1,1,0,0
10280338,0,1,0,1,1
12055855,0,0,1,0,0
12936739,0,1,0,1,0
'''
```

```python
df = pd.read_csv(StringIO(data),index_col = 0)
df
```

```python
df.shift(-1,axis = 1)     
```

```python
df1 = pd.concat([df,df.shift(-1,axis = 1)])
df1
```

```python
gp = df1.groupby(level = 0)
for k,v in gp:
    print(k,v)
```

```python
gp2 = df1.groupby('user_id')
for k,v in gp2:
    print(k,v)
```
