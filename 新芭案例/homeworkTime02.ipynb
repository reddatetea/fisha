```python
import numpy as np
import pandas as pd
import time
import datetime
```

```python
fname = r'f:\a00nutstore\堂堂博瑞\七年级八班学生作业完成时间统计表s.xlsx'
file_name = r'0915'
```

```python
df = pd.read_excel(fname,sheet_name = file_name)
df
```

```python
columns = df.columns.to_list()
num = columns[-1]
num
```

```python
# 格式化为时间元组
timeArray = time.localtime((num-25569) * 86400.0)
# 将时间元组转换为时间字符串
timeStr = time.strftime('%Y-%m-%d', timeArray)

print(timeStr)
```

```python
columns[-1] = timeStr
columns
```

```python
df = pd.read_excel(fname,sheet_name = file_name,names = columns)
df
```

```python
df[f'{timeStr}'] = df[f'{timeStr}'].astype('str').str[:5]
df
                 
```

```python
def getEndtime(timeStr):
    today = datetime.datetime.strptime(timeStr,'%Y-%m-%d')
    tomorrow = today + datetime.timedelta(days = 1)
    tomorrow_str = datetime.datetime.strftime(tomorrow,'%Y-%m-%d')
    print(tomorrow_str)
    return tomorrow_str
tomorrow_str = getEndtime(timeStr)
tomorrow_str
```

```python
df[f'{timeStr}'] = np.where(df[f'{timeStr}'].str[:2].isin(['00','01']),tomorrow_str + ' '+ df[f'{timeStr}']  ,timeStr + ' '+ df[f'{timeStr}'])
df
```
