```python
from io import StringIO
import pandas as pd
```

```python
data = '''
组别 值
A 1
A 2
A 3
B 4
B 5
C 6
C 7
B 8
'''
```

```python
df = pd.read_csv(StringIO(data),delim_whitespace = True)
df
```

```python
df.assign(所属组别的值 = df.groupby('组别').值.transform(lambda x : [x.to_list()]*len(x)))
```

```python
df.groupby('组别').值.agg(list)
```

```python
df.groupby('组别').值.agg(list).rename('所属组别的值')
```

```python
df.join(
     df.groupby('组别').值.agg(list).rename('所属组别的值'),
     on = '组别'
)
```

```python
#去除列表方括号
df.assign(所属组别的值 = df.groupby('组别').值.transform(lambda x : ','.join(x.astype(str))))
```
