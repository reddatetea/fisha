```python
import pandas as pd
import numpy as npaa
```

```python
index_arrays=[[1,1,2,2],
              ["男","女","男","女"]]
```

```python
colunms_arrays=[["2019","2019","2020","2020"],
                ["上半年","下半年","上半年","下半年"]]
```

```python
## 索引转换为多层
index=pd.MultiIndex.from_arrays(index_arrays,
                               names=("班级","性别"))
```

```python
## 列名转换为多层
columns=pd.MultiIndex.from_arrays(colunms_arrays,
                                 names=("年份","学期"))
```

```python
## 应用到DataFrame中
df=pd.DataFrame([(88,89,88,9),(77,89,97,98),
                (67,89,54,78),(34,67,89,54)],
               columns=columns,index=index)
df
```

```python
df.reset_index(inplace=True)
```

2019年上半年+2019年下半年

```python
df["2019全年合计"]=df[("2019","上半年")]+df[("2019","下半年")]
df
```

```python
df["2020全年合计"]=df[("2020","上半年")]+df[("2020","下半年")]
df
```
