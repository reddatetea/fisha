```python
import pandas as pd

df = pd.read_clipboard()
df
```

```python
def fun(x):
    if x >= 70 :
        return '优秀'
    elif x >= 60 :
        return '及格'
    else :
        return '加油'
```

```python
df1 = df.assign(total = df.loc[:,'Q1':'Q4'].sum(1),avg = df.loc[:,'Q1':'Q4'].mean(1))
df1 = df1.assign(review = df1.avg.map(fun))
df1 = df1.assign(name_first = df1.name.apply(lambda x:x[0]),name_len = df1.name.apply(lambda x:len(x)) )
df2 = df1.sort_values(['name_first','name_len']).iloc[:,:-2]
df2
```
