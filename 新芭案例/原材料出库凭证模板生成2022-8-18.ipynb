```python
dic = {'纸  小计':'5001002',
      '灰板  小计':'5001003',
      '辅料  小计':'5001004',
      '包装物  小计':'5001006'}
```

```python
def func(dic,s):
    s1 = []
    for i in s:
        j = dic.get(i.strip(),i)
        s1.append(j)
    return s1
```

```python
kemubianma = func(dic,df2['科目编码'].to_list())

kemubianma
```

```python
df_pinzhen['科目编码'] = kemubianma
df_pinzhen
```

```python
def jiner(kemu,jiner):
    jie = []
    dai = []
    for kemu,i in zip(kemu,jiner):
        if '计' in kemu:
            jie.append(i)
            dai.append(0)
        else :
            dai.append(i)
            jie.append(0)
    return jie,dai
    
    
```

```python
jie,dai = jiner(df2['科目编码'],df2['金额'])
jie,dai
```

```python
df_pinzhen['借方金额'] = jie
df_pinzhen['贷方金额'] = dai
```

```python
df_pinzhen['数量贷方'] = df2['出库'].to_list()
df_pinzhen['数量贷方']
```
