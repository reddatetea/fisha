```python
import pandas as pd
import numpy as np
```

```python
r = np.random.default_rng(666)
df = pd.DataFrame(r.integers(1, 20, size=[10, 5]),      #1-20之间的随机数，10行，5列
                  columns=[f'{i}月' for i in range(1, 6)]
                  )
df
```

```python
df1 = (
    df.assign(起始月=r.integers(1, 5, size=10))    #1-5之间的随机数，1行，10列
    .assign(月份数=r.integers(1, 4, size=10))      #1-4之间的随机数，1行，10列
)
df1
```

```python
#吉翾的作业解答2022-11-21
df1.assign(均值=df1.apply(lambda s: s.filter(regex='\d月$')
                .iloc[s.起始月-1: s.起始月-1+s.月份数]
                .mean(), axis=1)
          )

```
