```python
import pandas as pd
```

```python
df = pd.read_csv('20221117涨停数据.csv',encoding = 'gb18030')
df.head()
```

```python
df.columns
```

```python
(
 df[['代码','名称','涨停原因','首封','后封','连扳天','封单额']]
 .assign(涨停原因 = df.涨停原因.str.split('+'))  #转为列表，爆炸前准备
 .explode('涨停原因')
 .groupby('涨停原因').filter(lambda x:x.涨停原因.count() >= 3) #筛选出原因数大于3的数据
 .sort_values(by = ['后封','首封','连扳天','封单额'],ascending = [1,1,0,0])  #不同列分别排序
 .apply(lambda d : print(d.name,len(d),'\n',d,end = '\n\n'))
)
```
