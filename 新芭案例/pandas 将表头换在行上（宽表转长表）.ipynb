```python
import pandas as pd
from io import StringIO

data = '''
姓名,语文高一,语文高二,语文高三,数学高一,数学高二,数学高三
小明,32,48,97,55,98,54
小红,48,88,34,91,33,33
小李,14,6,97,41,70,75
小兰,82,35,45,15,4,44
'''
```

```python
df = pd.read_csv(StringIO(data))
df
```

```python
print(df.to_markdown())
```

```python
pd.wide_to_long(df, stubnames=['语文', '数学'],
                i='姓名',
                j='年级',
                suffix='\w+')
```
