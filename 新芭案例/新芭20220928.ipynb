```python
import pandas as pd
columns = ['车费', '午餐']
df1 = pd.DataFrame([(1, 2), (3, None)],
                   columns=columns,
                   index=['张三', '李四'])
df2 = pd.DataFrame([(None, 4), (7, 8)],
                   columns=columns,
                   index=['李四', '张三'])

df1
```

```python
df2
```

```python
df = df1.fillna(0) + df2.fillna(0)
df
```

```python
df['合计'] = df['车费'] + df['午餐']
df
```

```python
#链式
(
    df1.add(df2,fill_value= 0)
    .assign(合计 = lambda x:x.sum(1))
    .pipe(lambda x : pd.concat([x,x.sum().to_frame('合计').T]))
  )
```
