```python
import pandas as pd
df = pd.DataFrame({'品类':list('CABABDBA'),'数量':[10,20,30,50,10,40,50,90],'金额': list(range(100,108))})
df

```

```python
gp = df.groupby('品类').sum()
lei = pd.cut(gp.数量,bins = [0,40,80,100,1000],labels=[0.01,0.02,0.03,0.05])
df1= gp.assign(比例=list(lei))
df1
```

```python
dic = dict(zip(df1.index,df1.比例))
dic
```

```python
result = df.assign(提成=lambda x:x.品类.map(dic)*x.金额)
result
```

```python
result
```

```python
gp = df.groupby('品类').sum()
lei = pd.cut(gp.数量,bins = [0,20,40,60,1000],labels=[0.02,0.03,0.04,0.05])
result = gp.assign(比例=list(lei))
result = result.assign(提成 = lambda x:x.金额*x.比例)
result
```
