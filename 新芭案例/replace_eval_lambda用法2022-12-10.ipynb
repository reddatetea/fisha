```python
import pandas as pd
```

```python
df = pd.read_clipboard()
df
```

```python
foo = (
       df.原始数据
         .replace(r',','=',regex = True)
         .replace(r'\)=\(',',',regex = True)
        .map(lambda x:eval(f'dict{x}') if '(' in x else {})
)
foo
```

```python
(
 df.assign(foo = foo)
   .assign(bar= df.新增数据.str.split(','))
   .assign(total = lambda d:d.apply(lambda s:sum(s.foo[i] for i in s.bar if i in s.foo),axis = 1))
   )
```

```python
def getSum(s,d):
    x = s.foo
    if len(x) != 0:
        y = sum(x.values())
    else :
        y = 0
    return y
```

```python
df1 = df.assign(foo = foo).assign(bar= df.新增数据.str.split(','))
df1
```

```python
df1.assign(total = lambda x:x.apply(getSum,d=df1,axis = 1))
```
