```python
#作业# 增加一列，将 col0 列对应的索引值的列表显示在这个新列
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame(np.random.randint(0, 5, size=(10, 1)), columns=['col0'])
df
```

```python
df1 = df.copy()
```

```python
df2 = df1.reset_index()
df2.columns = ['col1','col0']
df2
```

```python
gp = df2.groupby('col0',sort = False).col1.apply(list)
df2  = pd.DataFrame(gp)
df2
```

```python
result = pd.merge(df1,df2,left_on = 'col0',right_index = True,how = 'left')
result
```
