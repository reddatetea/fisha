```python
'''
新芭作业https://wx.zsxq.com/dweb2/index/homework/812852112128442
'''
import pandas as pd
from io import StringIO

data = '''
Order_Numbers
44019541285_P_002 | 0317209757 | 87186978110350851 | 387186978103840544 | 0652569032
87186978110202440 | 44019119315 | 87186978110202440 | 44019119315
87186978110326832 | 44019453624 | 87186978110326832 | 44019453624
44019406029 | 0317196878 | 87186978110313085 | 387186978120481881|
44019480564 | 0317202711 | 87186978110335810 | 387186978103844160 | 630552498
'''

df = pd.read_csv(StringIO(data.replace(' ','')))
df
```

```python
df1 = df.Order_Numbers.str.split('|',expand=True)
df1
```

```python
df1 = df1.astype(str)
df1
```

```python
df2 = df1.where(lambda x:x.applymap(lambda y:y[:2]=='44'))           #where可以将不满足条件的标注为NaN
df2
```

```python
df3 = df2.bfill(axis = 1)    #bfill和backfill一样，是使用后一个有效值填充，由于axis=1，数据就向左填茺
df3
```

```python
df4 = df3[0]
df4
```

```python
df5 = df4.str.split('_').str.get(0)
df5
```
