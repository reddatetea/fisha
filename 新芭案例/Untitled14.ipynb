```python
import pandas as pd
import os
from fnmatch import fnmatch
```

```python
def qijianJisuan(string):
    y = int(string.split('-')[0])
    m = int(string.split('-')[1])
    d = int(string.split('-')[2])
    if d > 25 and m < 12:
        m = m + 1
    elif d > 25 and m == 12 :
        m = 1
        y = y + 1
    else :
        pass
    qijian = str(y)+'-'+ str(m).zfill(2)
    return qijian
```
