```python
import pandas as pd
import re

file = 'pandas正则分列示例.xlsx'
df = pd.read_excel(file)
df2 = df['车型名称'].map(lambda x: re.sub('[A-Z0-9. ]+', ',', x)).str.split(',', expand=True)
df = pd.concat([df, df2], axis=1)
df.columns = ['车型名称', '品牌', '车型']
print(df)
```
